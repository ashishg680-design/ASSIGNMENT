Q5  What are iterators in Python and how do they differ from iterables?
ans-In Python, iterators and iterables are related but distinct concepts.

Iterables:
- An iterable is an object that can return its members one at a time, allowing it to be iterated over in a loop (like a for loop).
- Examples of iterables include lists, tuples, sets, dictionaries, and strings.
- An iterable must implement the __iter__() method, which returns an iterator.

Iterators:
- An iterator is an object that allows you to iterate through an iterable.
- It keeps track of its current position and yields the next value when asked (next() is called).
- An iterator must implement the __next__() method (or next() in Python 2).
- Once an iterator is exhausted (no more items), it raises a StopIteration exception.

Key Differences:
- Iterables can be looped over multiple times. You can create multiple iterators from an iterable.
- Iterators are stateful and can typically be iterated over only once. After exhaustion, they can't be reused.

Example:
my_list = [1, 2, 3]  # Iterable
my_iter = iter(my_list)  # Iterator

print(next(my_iter))  # 1
print(next(my_iter))  # 2
print(next(my_iter))  # 3
next(my_iter) would raise StopIteration

Q6 . Explain the concept of generators in Python and how they are defined.?
ans-Generators in Python are a type of iterable, like lists or tuples. However, unlike lists, generators don't store all the values in memory at once. Instead, they generate values on the fly as they're needed.

How Generators Are Defined:
- *Using a Function with yield*: Generators are defined using a function where instead of using return, you use yield to produce a value. When a generator is iterated over, it starts executing the function until it hits a yield, then it returns that value and pauses. When asked for the next value, it resumes execution until the next yield.

Example:
def my_generator():
    yield 1
    yield 2
    yield 3

for value in my_generator():
    print(value)
This will print 1, 2, and 3.

Q8 What is a lambda function in Python and when is it typically used?
ans-In Python, a lambda function is a small anonymous function defined with the lambda keyword. It's used for short, one-time operations where defining a full-fledged function isn't necessary.

Lambda functions are typically used for:
- Short operations like data transformations or filtering in functions like map(), filter(), or sorted().
- When you need a quick function without declaring a named function.

Example:
Using lambda with map to square numbers
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)  # Output: [1, 4, 9, 16]

Q10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans-In Python, map(), reduce(), and filter() are three higher-order functions that operate on iterables. Here's how they differ:

- *map(function, iterable):*
    - Applies a given function to each item of an iterable (like a list or tuple).
    - Returns a map object (an iterator) yielding the results.
    - Example: list(map(lambda x: x**2, [1, 2, 3])) gives [1, 4, 9].
- *filter(function, iterable):*
    - Applies a function to each item of an iterable to decide whether to include it in the output.
    - Returns a filter object (an iterator) with items for which the function returns True.
    - Example: list(filter(lambda x: x > 2, [1, 2, 3, 4])) gives [3, 4].
- *reduce(function, iterable):*
    - In Python 3, reduce() is moved to the functools module, so you need from functools import reduce.
    - Applies a function of two arguments cumulatively to the items of an iterable to reduce it to a single value.
    - Example: reduce(lambda x, y: x + y, [1, 2, 3, 4]) gives 10

```python
# Q1 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
# the list.
def sum_of_evens(numbers):
  total=0
  for num in numbers:
    if num % 2==0:
      # check if number is even
      total+=num
  return total

my_list=[1,2,3,4,5,6,7]
print("sum of ven numbers:",sum_of_evens(my_list))
```

```python
# Q2. Create a Python function that accepts a string and returns the reverse of that string.?
def reverse_string(input_str):
  reversed_string=""
  for char in input_str:
    reversed_string=char+reversed_string
  return reversed_string

my_string="hello"
print("reversed string:",reverse_string(my_string))
```

```python
# Q6  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(n):
  for i in range(n+1):
    yield 2**i
for power in power_of_two(5):
  print(power)
```

```python
# Q7 7. Implement a generator function that reads a file line by line and yields each line as a string.
# Create a dummy file named myfile.txt for demonstration
with open("myfile.txt", "w") as f:
    f.write("My name in ashish.\n")
    f.write("I am a working professtional.\n")
    f.write("In present i am done my Data Analysis course.")

    # Generator function
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()  # yield each line without extra \n

# Example usage:
for line in read_file_line_by_line("myfile.txt"):
    print(line)
```

```python
# Q10 Create a Python program that uses `filter()` to remove all the vowels from a given string.?
# Program to remove vowels using filter()

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, s))

# Example
text = "Hello World"
result = remove_vowels(text)
print("Original:", text)
print("Without vowels:", result)
```
